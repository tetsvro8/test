```python
class Adnw:
    def __init__(self, id, price = None):
        self.id = id
        self.price = []
    
    def set_price(self, lst):
        tmp = lst[lst['winner_id']==self.id]['winner_bid_price_adv_currency']
        tm = tmp.as_matrix()
        for i in range(tmp.count()):
            #print(tmp.as_matrix()[i])
            self.price.append(tm[i])

adnw_lst = set()
price_lst = {}
count_lst = {}
hit_or_miss_lst = {}

get = 0;

#データ　セット　初期化
for i in set((lst['winner_id'])):
    adnw = Adnw(i)
    adnw.set_price(lst)
    
    adnw_lst.add(adnw.id)
    price_lst[i] = adnw.price
    #print(i)
    count_lst[i] = 0
    hit_or_miss_lst[i] = 0.0
    
    #print(adnw.price)
    #print(price_lst[i][len(price_lst[i])-1])

print('normal get')
print(sum(lst['winner_bid_price_adv_currency']))

adnw_act = {}

def trial(try_num, adnw_lst, price_lst, count_lst, hit_or_miss_lst):
    get = 0.0
    
    for i in range(try_num):
        j = random.choice(list(adnw_lst),1)[0]
        #print(j)
        #print(count_lst[j])
        get += price_lst[j][count_lst[j]%len(price_lst[j])]
        adnw_act[j].append(price_lst[j][count_lst[j]%len(price_lst[j])])
        #print(j)
        #print(adnw_act[j])
        count_lst[j] += 1
        hit_or_miss_lst[j] = count_lst[j] / (i + 1)
    return get

print('rondom get')
try_num = 100
print(trial(try_num, adnw_lst, price_lst, count_lst, hit_or_miss_lst))

print("kari")
print(adnw_lst)
print("jissai")
#print(adnw_act[2][4])
print("count")
print(count_lst)

"""
sum_count = 0

for j in set(adnw_lst):
    sum_count += count_lst[j]

print(sum_count)
"""

argmax(adnw_act)
adnw_act
```
